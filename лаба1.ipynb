# Задание 1

```python
def sigmoid(x):
    return 1.0 / (1 + np.exp(-x))

x = np.linspace(-5, 5, 100)
y = sigmoid(x)

plt.plot(x, y)
plt.ylim(0, 1)
plt.show()
```

# Задание 2

```python
import numpy as np
import matplotlib.pyplot as plt

v = np.array([0,1])
a = 5
M = np.array([
              [math.cos(a), -math.sin(a)],
              [math.sin(a), math.cos(a)]
])
vs = [v]
for i in range(100):
    v = vs[i] @ M * 0.99
    vs.append(v)
    
vs = np.array(vs)

plt.figure(figsize=(6, 6))
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.plot(*vs.T)
plt.show()
```

```python
def calculate_row(x, p):
  return list(map(lambda power: x ** power, range(p + 1)))

print(calculate_row(2, 5))
```

# Задание 4


```python
def rastrigin(x, y):
    return 20 + (x ** 2 - 10 * np.cos(2 * math.pi * x)) + (y ** 2 - 10 * np.cos(2 * math.pi * y))

x = np.linspace(-5, 5, 1000)[:, np.newaxis]
y = np.linspace(-5, 5, 1000)[:, np.newaxis]
X, Y = np.meshgrid(x, y)

plt.figure(figsize=(10, 10))
plt.contourf(rastrigin(X, Y), levels=10)
plt.show()
```
